```python
import cv2  
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
image = cv2.imread('full_image.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

pattern = cv2.imread('pattern.jpg')
pattern_gray = cv2.cvtColor(pattern, cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.figure()
plt.imshow(cv2.cvtColor(pattern, cv2.COLOR_BGR2RGB))
```

```python
simple_match = cv2.matchTemplate(image_gray, pattern_gray, cv2.TM_CCORR_NORMED)

sns.heatmap(simple_match)
```

```python
i, j = np.unravel_index(simple_match.argmax(), simple_match.shape)

w, h = pattern.shape[:-1]
plot_image = image.copy()

cv2.rectangle(plot_image, (j, i), (j + h, i + w), (255, 0, 0), 8)

plt.imshow(cv2.cvtColor(plot_image, cv2.COLOR_BGR2RGB))
```

```python
i, j = np.unravel_index(simple_match.argmax(), simple_match.shape)

plt.imshow(cv2.cvtColor(image[i:i+w,j:j+h], cv2.COLOR_BGR2RGB))
```
