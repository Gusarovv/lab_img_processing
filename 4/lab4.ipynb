Цель лабораторной работы:
Преобразование изображение для последующего перевода

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
img = cv.imread('lab4.jpg', cv.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
ret, thresh = cv.threshold(img, 110, 255, cv.THRESH_BINARY_INV + cv.THRESH_BINARY)

plt.figure(figsize=(16,10))
plt.imshow(thresh, 'gray')
```

```python
ret, thresh = cv.threshold(img, 100, 255, cv.THRESH_BINARY)

plt.figure(figsize=(16,10))
plt.imshow(thresh, 'gray')
```

```python
thresh2 = cv.erode(thresh, np.ones((2,2), np.int8))

plt.figure(figsize=(16,10))
plt.imshow(thresh2, 'gray')
```
